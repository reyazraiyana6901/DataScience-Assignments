```python
data["Y"] = data["Y"].map({"M": 0, "R": 1})

# Verify conversion
print("\nUnique values after encoding:", data["Y"].unique())
```

```python
X = data.drop(columns=["Y"])
y = data["Y"]
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_scaled = sc.fit_transform(X)

print("\nScaled feature matrix shape:", X_scaled.shape)
```

# 2. Model Implementation

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y,test_size=0.25,random_state=42,stratify=y)

print("\nTrain shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

tf.random.set_seed(42)

model = Sequential()

# Input layer + 1 hidden layer
model.add(Dense(units=30,activation='relu',input_dim=X_train.shape[1]))

# Output layer (binary)
model.add(Dense(units=1,activation='sigmoid'))

# Display model summary
print("\n===== ANN MODEL SUMMARY =====")
model.summary()
```

```python
model.compile( optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,epochs=50,batch_size=20,validation_split=0.2,verbose=1)
```

```python
# Predict probabilities
y_pred_proba = model.predict(X_test)

# Convert to 0/1
y_pred = (y_pred_proba > 0.5).astype(int)

print("\nFirst 10 predictions:", y_pred[:10].flatten())
```

# 3. Hyperparameter Tuning

```python
import random

param_dist = {
    "neurons": [20, 30, 40, 50],
    "activation": ['relu', 'tanh'],
    "learning_rate": [0.001, 0.005, 0.01],
    "batch_size": [10, 20, 30],
    "epochs": [30, 50, 70]
}
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score

results = []
n_iter = 10   # Number of random combinations

for i in range(n_iter):
    print(f"\n===== Running iteration {i+1}/{n_iter} =====")
    
    # Pick random hyperparameters
    neurons = random.choice(param_dist["neurons"])
    activation = random.choice(param_dist["activation"])
    lr = random.choice(param_dist["learning_rate"])
    batch_size = random.choice(param_dist["batch_size"])
    epochs = random.choice(param_dist["epochs"])
    
    # Build model
    model = Sequential()
    model.add(Dense(neurons, activation=activation, input_dim=X_train.shape[1]))
    model.add(Dense(neurons//2, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    # Train
    model.fit(X_train, y_train,
              batch_size=batch_size,
              epochs=epochs,
              validation_split=0.2,
              verbose=0)
    
    # Evaluate on test set
    y_pred = (model.predict(X_test) > 0.5).astype(int)
    accuracy = accuracy_score(y_test, y_pred)
    
    # Save results
    results.append({
        "neurons": neurons,
        "activation": activation,
        "learning_rate": lr,
        "batch_size": batch_size,
        "epochs": epochs,
        "accuracy": accuracy
    })
    
    print("Accuracy:", accuracy)
```

```python
best_model = max(results, key=lambda x: x["accuracy"])
print("\n===== BEST HYPERPARAMETERS FOUND =====")
print(best_model)
```

# 4. Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Predict probabilities
y_pred_proba = model.predict(X_test)

# Convert to class labels (0/1)
y_pred = (y_pred_proba > 0.5).astype(int)


# EVALUATION METRICS

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("\n===== MODEL PERFORMANCE =====")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)

print("\n===== CONFUSION MATRIX =====")
print(cm)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

best_params = {
    'neurons': 20,
    'activation': 'relu',
    'learning_rate': 0.001,
    'batch_size': 30,
    'epochs': 50
}

tuned_model = Sequential()
tuned_model.add(Dense(best_params['neurons'], activation=best_params['activation'], input_dim=X_train.shape[1]))
tuned_model.add(Dense(best_params['neurons']//2, activation=best_params['activation']))
tuned_model.add(Dense(1, activation='sigmoid'))

tuned_model.compile(optimizer=Adam(learning_rate=best_params['learning_rate']),
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

tuned_model.fit(X_train, y_train,
                epochs=best_params['epochs'],
                batch_size=best_params['batch_size'],
                validation_split=0.2,
                verbose=0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred_tuned = (tuned_model.predict(X_test) > 0.5).astype(int)

tuned_acc = accuracy_score(y_test, y_pred_tuned)
tuned_prec = precision_score(y_test, y_pred_tuned)
tuned_rec  = recall_score(y_test, y_pred_tuned)
tuned_f1   = f1_score(y_test, y_pred_tuned)

print("\n===== TUNED MODEL PERFORMANCE =====")
print("Accuracy :", tuned_acc)
print("Precision:", tuned_prec)
print("Recall   :", tuned_rec)
print("F1-score :", tuned_f1)
```
