1) Standard Scaling (Z-score):

- Not ideal for When outliers exist (they affect mean & SD)
2) Min-Max Scaling:
- Useful for Neural Networks, KNN, Decision Trees, or when using distance-based algorithms (to avoid one feature dominating due to larger scale)
- Not ideal for If outliers exist — they shrink everything else

#### ONE HOT ENCODING
1) Pros:
- No ordinal assumptions – treats all categories as independent.
- Works well for linear models (Logistic Regression, Linear Regression) that are sensitive to numeric magnitude.
- Easy to interpret – each dummy column clearly represents one category.
2) Cons:
- High dimensionality – if a categorical column has many categories, it creates many columns (curse of dimensionality).
- Memory inefficient – large datasets with many categories can become very big.
- May need drop_first=True to avoid multicollinearity in linear models.


#### LABEL ENCODING
1) Pros:
- Memory efficient – only replaces column with integers, no extra columns.
- Fast to implement – works well for high-cardinality categorical variables.
2) Cons:
- Implied ordinality – integers may mislead algorithms that interpret numeric order (e.g., Linear Regression, SVM).

## FEATURE ENGINEERING

#### Net Capital:
- People may have high income if they have high net capital gains, even if they work fewer hours.
- Combining capital_gain and capital_loss into a single numeric feature can give the model a better signal about total financial gains.

```python
df['net_capital'] = df['capital_gain'] - df['capital_loss']
```

### Workload Category (Overworked / Normal / Underworked):
- Based on: hours_per_week
- People working very high hours may have higher income, and those working fewer hours may have lower income.

```python
def workload_category(hours):
    if hours < 35:
        return 'underworked'
    elif hours <= 50:
        return 'normal'
    else:
        return 'overworked'

df['workload_category'] = df['hours_per_week'].apply(workload_category)
```

```python
# Some of these, especially capital_gain and capital_loss, are typically highly skewed (most people have 0, few have very large values).
skewed_features = df[['capital_gain', 'capital_loss', 'hours_per_week']].skew()
print(skewed_features)
```

- Highly skewed features → capital_gain and capital_loss
- Nearly symmetric → hours_per_week (no transformation needed)

```python
df['capital_gain_log'] = np.log1p(df['capital_gain'])
df['capital_loss_log'] = np.log1p(df['capital_loss'])
```

```python
print(df[['capital_gain_log', 'capital_loss_log']].skew())
''' Log transformation significantly reduced extreme right skew '''
```
