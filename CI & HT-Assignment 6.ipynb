import numpy as np
import math
from scipy import stats

## CI

```python
data = [1.13,1.55,1.43,0.92,1.25,1.36,1.32,0.85,1.07,1.48,1.20,1.33,1.18,1.22,1.29]
```

#### a. Build 99% Confidence Interval Using Sample Standard Deviation
We use ttest as sample size is also less than 30 and population mean is unknown.

```python
n = len(data)
mean = np.mean(data)
s = np.std(data, ddof=1)
t_val = stats.t.ppf(0.995, df=n-1)
s1 = s/math.sqrt(n)
margin =t_val*s1
ci = (m-margin, m+margin)
print("I am 99% confident that the true mean durability of the print-head lies in: ",ci)
```

#### b. Build 99% Confidence Interval Using Known Population Standard Deviation
the population standard deviation is 0.2 million characters so we can use ztest

```python
sigma = 0.2
z_val = stats.norm.ppf(0.995)  # two-tailed 99%
SE = sigma / math.sqrt(n) # Standard error
ME = z_val * SE # Margin of error
# Confidence interval
CI = (mean - ME, mean + ME)

print("99% CI using known sigma:", CI)
```

## HT

1) Null Hypothesis (H0): mean weekly cost follows the theoretical model and
 Alternative Hypothesis (H1): mean weekly cost is higher than the theoretical model
2) Calculate the Test Statistic:Since the population standard deviation is known and the sample is from a normal distribution, we can use a z-test.

```python
# Given data
sample_mean = 3050        
theoretical_mean = 1000 + 5*600
sigma = 5 * 25           
n = 25
```

```python
from statsmodels.stats.weightstats import ztest
sample_data = np.random.normal(loc=sample_mean, scale=sigma, size=n)
# Perform z-test
z_stat, p_value = ztest(sample_data, value=theoretical_mean, alternative='larger')
alpha = 0.05
print("Z-test statistic:", z_stat)
```

4. Make a Decision:


```python
 alpha = 0.05
print("P-value:", p_value)
if p_value < alpha:
    print("Reject H0: Evidence suggests costs are higher than predicted.")
else:
    print("Fail to reject H0: No evidence that costs are higher than predicted.")
```

5) Conclusion:

There is no statistical evidence that the weekly operating costs are higher than predicted by the model.
